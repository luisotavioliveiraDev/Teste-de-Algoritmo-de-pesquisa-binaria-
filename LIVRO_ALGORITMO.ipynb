PESQUISA BINARIA: algoritmo de busca (IA, bancos de dados etc)
  
   Exemplos : procurar um numero em uma lista telefonica, acertar um numero de 1
   a 100 A pesquisa Binaria precisa de Log2n para retornar o valor correto (OBS:Pesquisa ordenaria só funciona em lista ordenadas)
  
  - Codigo de Exemplo

```python

def pesquisa_binaria(lista, item):
    baixo = 0
    alto = len(lista) - 1

    while baixo <= alto:
        meio = (baixo+alto) // 2
        chute = lista[meio]
        if chute == item:
          return meio
        if chute > item:
          alto = meio -1
        else:
          baixo = meio + 1
    return None
Minha_lista =[1,3,5,7,9]
print(pesquisa_binaria(Minha_lista,9))
```

Notação Big 0 : notação especial que diz quão rápido é um algoritmo

- Tempo de execução dos agoritmos cresce a taxas diferentes

- o tempo na notação Big O é O(n)

- na notação Big O não tem tempo ela compara o numero de oprações e permite ver quão rapido um algoritmo cresce

- exemplo: suponha que você tenha que desenhar uma grade com 16 divisões

  ALGORITMO 1: desenhando as divisões uma por uma nesse algoritmo você tem que desenhar 16 divisões ou seja sendo assim vai passar por 16 etapas

  tendo em vista que a notação Big O conta a quantidade de operações esse algorimo quantidade de divisões = quantidade de operações

  ALGORITMO 2 : dobrando o papel uam vez faz uma operação e duas divisões dobrando o papel a cada vez que dobra o papel as divisões duplicam fazem 16 divisões em 4 etapas

  logo no segundo algoritmo a quantidade de operações seria operação = 2 divisões

Resumo :

 - A pesquisa binaria é muito mais rapida do que pesquisa simples

 - O(log n) é mais raido que O(n) e 0(log n) fica ainda mais rapido a medida que os elementos aumentam

 - A rapidez de um algoritmo não é medida em segundos e sim em operações quantos menos operações mais rapido um algoritmo

 - tempo de execução de um algoritmo é medido por meio de seu crescimento

 - o tempo de execução dos algoritmos é expresso na notação Big O
